```python
# import all modules necessary
import numpy as np 
import seaborn as sns
from numpy.random import normal
from datetime import datetime
import matplotlib.pyplot as plt
import math
import pandas as pd
```

```python
data = pd.read_csv('Falltest2.csv') # We read the data using pandas
time1 = (data['time'])
ax = (data['ax'])
ay = (data['ay'])
az = (data['az'])
atotal = (data['atotal'])

```

```python
fig3 = plt.figure(figsize=(4,4) )
ax3= fig3.add_subplot(projection='3d')

ax3.scatter(np.log(ax), np.log(ay), np.log(az), c=az, cmap='autumn', linewidth=0.4, edgecolor='gray')



ax3.set_xlabel('ax ($m/s^2$)')
ax3.set_ylabel('ay ($m/s^2$)')
ax3.set_zlabel('az ($m/s^2$)')

ax3.azim = 15
ax3.dist = 10
ax3.elev = 35
plt.title('Plot Of Components Of Acceleration Of 2nd Phone Drop', weight='bold')
plt.show()

fig3.savefig('3dfallphone.png', bbox_inches='tight')

```
