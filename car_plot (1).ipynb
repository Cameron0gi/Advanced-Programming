```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import pandas as pd
import statistics
```

```python
data = pd.read_csv('DriveTest.csv') # We read the data using pandas
time = (data['time'])  #Read time into variable
atotal = (data['atotal'])  #Read total accelaration into variable
```

```python
maxt = time.max()
newt = [round(i) for i in time]  #Round all time values

 
```

```python
def createav(x):  
    i = 0
    average = []
    while i < len(x):
        i = i + 100
    
        average.append(((np.average(x[i:i+100]))))
    return(average)
```

```python
time = createav(newt)  #Create average for time
totalacc = createav(atotal) #create average for accelaration
```

```python
fig = plt.figure(figsize=(10,8))  #Create figure to plot on 

plt.plot(time, totalacc, color='b')
plt.xlabel('Time (s)', fontsize=15)
plt.ylabel('Average Acceleration ($m/s^2$)', fontsize=15)
plt.title('Accelaration on a 400m Car Journey', fontsize=15, weight='bold')

fig.savefig('car_journey.svg')
```
